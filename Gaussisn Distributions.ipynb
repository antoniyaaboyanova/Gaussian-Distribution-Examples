**Visualisation of a Gaussian Distribution**

```python
#Packages 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

**Standard Normal Distribution**

```python
#Standard Dist Parameters 
mu = 0
var = 1
sigma = np.sqrt(var)
x = np.linspace(-8, 8, 100)


#Plotting
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title('SND')
plt.xlabel('x')
plt.ylabel('pdf (x)')
plt.show()
```
